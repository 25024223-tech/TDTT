# W5A3 Viết hàm kiểm tra một số có phải là số nguyên tố hay khô

```python
import math
def la_snt(n):
    if n < 2:
        return False
    else:
        for i in range (2, math.isqrt(n)+1):
            if n % i == 0:
                return False
        return True
n = int(input())
print(la_snt(n))
```

# W5A6 Viết hàm tính giai thừa. Nhập vào số nguyên n, trả về giai thừa của n (n!) 

```python
def factorial(n):
    tich=1
    if n>=0:
        for i in range(1,n+1):
            tich *=i
    return tich
    
n=int(input())
print(factorial(n))

```

# W5A14 
Viết hàm compute(a, b, op) thực hiện phép tính a (+|-|*|/) b tùy theo ký tự op. 

```python
def compute(a, b, op):
    if op == "+":
        return a + b
    elif op == "-":
        return a - b
    elif op == "*":
        return a * b
    elif op == "/":
        return a / b
a, b = map(int, input().split())
op = input()
print(compute(a, b, op))
```

# W5A15
 Viết hàm is_right_triangle(a, b, c) trả về True nếu ba số có thể là ba cạnh của tam giác vuông

```python
def is_right_triangle(a,b,c):
    if a**2 == c**2 +b**2 or b**2 == a**2 + c**2 or c**2 == a**2 + b**2:
        return True
    return False
a,b,c = map(int,input().split())
print(is_right_triangle(a,b,c))
```
